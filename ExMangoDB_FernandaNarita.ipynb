```python
from pymongo import MongoClient
import pandas as pd 
```

```python
client = MongoClient("mongodb://localhost:27017/")
```

```python
df = pd.read_csv("HousesInfo.csv")
```

```python
df.head()
```

```python
df.columns = ["qte_banheiros", "qte_quartos", "area", "cep", "preco"]
df
```

```python
minha_lista = df.to_dict("records")
minha_lista
```

```python
meu_banco = client["cadastro"]
houses = meu_banco["houses"]
_ids = houses.insert_many(minha_lista)
```

```python
print(_ids.inserted_ids)
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

df = pd.DataFrame(mycol.find())
df
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

x = mycol.find({'$and':[{'area': {"$lt":1000}}, {"area": {"$gt": 200}}]})

df=pd.DataFrame(x)
df
```

```python
import matplotlib.pyplot as plt
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

mydocs = mycol.find({'$and':[{'area': {"$lt":1000}}, {"area": {"$gt": 200}}]})

df=pd.DataFrame(mydocs)
grafico1 = df.plot.scatter(x = "area", y = "preco")
plt.show()
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

mydoc = mycol.find().sort("preco", -1).limit(15)

df=pd.DataFrame(mydoc)
grafico2 = df.plot.line(x = "preco", y = "qte_quartos")
plt.show()
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

myquery = {"qte_banheiros": 3}
mydoc = mycol.find(myquery)

df = pd.DataFrame(mydoc)
grafico3 = df.plot.bar(x = "qte_banheiros", y = "qte_quartos")
plt.show()
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

myquery = {"cep": {"$gt":50000} }
newvalues = {"$set": {"preco": "1000"}}

x = mycol.update_many(myquery, newvalues)

print (x.modified_count, "Docs atualizados")
```

```python
mydb= client["cadastro"]
mycol = mydb["houses"]

myquery = {"cep": {"$gt": 40000}}

x =  mycol.delete_many(myquery)
           
print(x.deleted_count, "Docs deleted")
```
